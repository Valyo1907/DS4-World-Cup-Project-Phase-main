```python
# Most frequent matchups
matchup_counts = df_hist["match_name"].replace("", np.nan).dropna().value_counts().reset_index()
matchup_counts.columns = ["match_name", "matches"]
matchup_counts
```

```python
# Matchups with order from most frequent matchups
matchup_counts = df_hist["match_name"].replace("", np.nan).dropna().value_counts().reset_index()
matchup_counts.columns = ["match_name", "matches"]

order_by_frequency = matchup_counts.head(20)["match_name"].tolist()
order_by_frequency
```

```python
# Build wins/draws summary aligned to the most frequent matchups
# Use goals to determine result so totals match overall match counts
matchups_df = df_hist.copy()
matchups_df = matchups_df[matchups_df["match_name"].notna() & (matchups_df["match_name"] != "")]
matchups_df
```

```python
# Check goals and missing values used to determine outcomes
home_goals = pd.to_numeric(matchups_df["home_goals"], errors="coerce")
away_goals = pd.to_numeric(matchups_df["away_goals"], errors="coerce")
missing_goals = home_goals.isna() | away_goals.isna()

(
    matchups_df[["match_name", "home_goals", "away_goals", "home_team_win", "away_team_win", "draw"]]
    .assign(missing_goals=missing_goals)
    .head(10)
)
```

```python
# Compute outcomes per match (home win / away win / draw)
outcome = np.where(
    home_goals.gt(away_goals),
    "home_wins",
    np.where(home_goals.lt(away_goals), "away_wins", "draws"),
)

# Fallback to flags when goals are missing
if missing_goals.any():
    outcome = pd.Series(outcome, index=matchups_df.index)
    outcome.loc[missing_goals & (matchups_df["home_team_win"] == 1)] = "home_wins"
    outcome.loc[missing_goals & (matchups_df["away_team_win"] == 1)] = "away_wins"
    outcome.loc[missing_goals & (matchups_df["draw"] == 1)] = "draws"
    outcome = outcome.fillna("unknown")
else:
    outcome = pd.Series(outcome, index=matchups_df.index)

matchups_df = matchups_df.assign(outcome=outcome)
matchups_df["outcome"].value_counts()
```

```python
# Aggregate outcomes per matchup
matchup_summary = (
    matchups_df.pivot_table(
        index="match_name",
        columns="outcome",
        values="year",
        aggfunc="count",
        fill_value=0,
    )
    .reset_index()
)

# Ensure required columns exist
for col in ["home_wins", "away_wins", "draws", "unknown"]:
    if col not in matchup_summary.columns:
        matchup_summary[col] = 0

matchup_summary.head(10)
```

```python
# Align to the most frequent matchups and compute totals
matchup_summary = matchup_counts.merge(matchup_summary, on="match_name", how="left").fillna(0)
matchup_summary = matchup_summary[matchup_summary["match_name"].isin(order_by_frequency)]
matchup_summary["total_results"] = (
    matchup_summary["home_wins"]
    + matchup_summary["away_wins"]
    + matchup_summary["draws"]
    + matchup_summary["unknown"]
)

matchup_summary.sort_values("matches", ascending=False).head(20)
```

```python
# Table shows totals aligned to the most frequent matchups
matchup_summary = matchup_summary.sort_values("matches", ascending=False)
matchup_summary
```

```python
fig = px.bar(
    matchup_summary,
    x="match_name",
    y=["home_wins", "away_wins", "draws", "unknown"],
    title="Most Frequent Matchups by Result (Home/Away/Draw)",
    barmode="stack",
    category_orders={"match_name": order_by_frequency},
)
fig

# Same data, sorted by home wins
matchup_win_sorted = matchup_summary.sort_values("home_wins", ascending=False)

fig = px.bar(
    matchup_win_sorted,
    x="match_name",
    y=["home_wins", "away_wins", "draws", "unknown"],
    title="Matchups Sorted by Home Wins (Home/Away/Draw)",
    barmode="stack",
)
fig
```
